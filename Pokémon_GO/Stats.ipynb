```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lvl_data = pd.read_csv('XP.txt', sep='\s+', header=None, skiprows=0)
data = lvl_data.rename(columns = {0:'lvl',1: 'xp',2:'xp_tot'})
lvl = (data['lvl'].values.tolist())
xp = (data['xp'].values.tolist())
xp_tot = (data['xp_tot'].values.tolist())
#print(lvl)
#print(xp)
#print(xp_tot)
```

```python
plt.plot(lvl, xp, 'r')
plt.xlabel('Level')
plt.xlim(1, 50)
plt.xticks(np.arange(5, 51, 5)) 
plt.ylabel('Required XP')
plt.ylim(1000, 30000000)
plt.tick_params(axis = 'y', rotation =-45)
plt.title('Pokémon GO XP Curve (Linear Plot)')
plt.show()

plt.plot(lvl, xp, 'r')
plt.xlabel('Level')
plt.xlim(1, 50)
plt.xticks(np.arange(5, 51, 5)) 
plt.ylabel('Required XP')
plt.ylim(1000, 30000000)
plt.yscale('log')
plt.tick_params(axis = 'y', rotation =-45)
plt.title('Pokémon GO XP Curve (Log Plot)')
plt.show()
```

```python
#lvl 41 requires 6,000,000 xp
earlygame = []
lategame = []
for i in xp:
    if i >= 6000000:
        lategame.append(i)
for i in xp:
    if i < 6000000:
        earlygame.append(i)
        
sum = 0
for i in range(0, len(earlygame)):    
    sum = sum + earlygame[i]; 
e = sum
sum = 0
for i in range(0, len(lategame)):    
    sum = sum + lategame[i]; 
l = sum
ratio = l/e
#print("The total xp required to reach level 40 is %s." %(e))
#print("The total xp required to go from level 40 to 50 is %s." %(l))
print("The total xp required to go from level 40 to 50 is %s times more than the total xp required to reach level 40." %(ratio))
```
