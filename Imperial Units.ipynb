```python
#List of units converting from
s = ["thous", "inches", "feet", "yards", "chains", "furlongs", "miles", "leagues"]
print("\033[1m" + "List of supported units:" + "\033[0m")
print("thous")
print("inches")
print("feet")
print("yards")
print("chains")
print("furlongs")
print("miles")
print("leagues")
unit = False
while unit == False:
    unit = input("Units converting from? ")
    if unit == "thous":
        t = int(input())
        i = t/(1000)
        f = t/(1000*12)
        y = t/(1000*12*3)
        c = t/(1000*12*3*22)
        u = t/(1000*12*3*22*10)
        m = t/(1000*12*3*22*10*8)
        l = t/(1000*12*3*22*10*8*3)
        print(t, "tous is:")
        print(i, "inches")
        print(f, "feet")
        print(y, "yards")
        print(c, "chains")
        print(u, "furlongs")
        print(m, "miles")
        print(l, "leagues")
    elif unit == "inches":
        i = int(input())
        t = i*(1000)
        f = i/(12)
        y = i/(12*3)
        c = i/(12*3*22)
        u = i/(12*3*22*10)
        m = i/(12*3*22*10*8)
        l = i/(12*3*22*10*8*3)
        print(i, "inches is:")
        print(t, "tous")
        print(f, "feet")
        print(y, "yards")
        print(c, "chains")
        print(u, "furlongs")
        print(m, "miles")
        print(l, "leagues")
    elif unit == "feet":
        f = int(input())
        t = f*(1000*12)
        i = f*(12)
        y = f/(3)
        c = f/(3*22)
        u = f/(3*22*10)
        m = f/(3*22*10*8)
        l = f/(3*22*10*8*3)
        print(f, "feet is:")
        print(t, "tous")
        print(i, "inches")
        print(y, "yards")
        print(c, "chains")
        print(u, "furlongs")
        print(m, "miles")
        print(l, "leagues")
    elif unit == "yards":
        y = int(input())
        t = y*(1000*12*3)
        i = y*(12*3)
        f = y*(3)
        c = y/(22)
        u = y/(22*10)
        m = y/(22*10*8)
        l = y/(22*10*8*3)
        print(y, "yards is:")
        print(t, "tous")
        print(i, "inches")
        print(f, "feet")
        print(c, "chains")
        print(u, "furlongs")
        print(m, "miles")
        print(l, "leagues")
    elif unit == "chains":
        c = int(input())
        t = c*(1000*12*3*22)
        i = c*(12*3*22)
        f = c*(3*22)
        y = c*(22)
        u = c/(10)
        m = c/(10*8)
        l = c/(10*8*3)
        print(c, "chains is:")
        print(t, "tous")
        print(i, "inches")
        print(f, "feet")
        print(y, "yards")
        print(u, "furlongs")
        print(m, "miles")
        print(l, "leagues")
    elif unit == "furlongs":
        u = int(input())
        t = u*(1000*12*3*22*10)
        i = u*(12*3*22*10)
        f = u*(3*22*10)
        y = u*(22*10)
        c = u*(10)
        m = u/(8)
        l = u/(8*3)
        print(u, "furlongs is:")
        print(t, "tous")
        print(i, "inches")
        print(f, "feet")
        print(y, "yards")
        print(c, "chains")
        print(m, "miles")
        print(l, "leagues")
    elif unit == "miles":
        m = int(input())
        t = m*(1000*12*3*22*10*8)
        i = m*(12*3*22*10*8)
        f = m*(3*22*10*8)
        y = m*(22*10*8)
        c = m*(10*8)
        u = m*(8)
        l = m/(3)
        print(m, "miles is:")
        print(t, "tous")
        print(i, "inches")
        print(f, "feet")
        print(y, "yards")
        print(c, "chains")
        print(u, "furlongs")
        print(l, "leagues")
    elif unit == "leagues":
        l = int(input())
        t = l*(1000*12*3*22*10*8*3)
        i = l*(12*3*22*10*8*3)
        f = l*(3*22*10*8*3)
        y = l*(22*10*8*3)
        c = l*(10*8*3)
        u = l*(8*3)
        m = l*(3)
        print(l, "leagues is:")
        print(t, "tous")
        print(i, "inches")
        print(f, "feet")
        print(y, "yards")
        print(c, "chains")
        print(u, "furlongs")
        print(m, "miles")
    else:
        print("Unit not yet supported!")
```
